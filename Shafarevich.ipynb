## Shafarevich's Theorem on Solvable Galois Groups

The theorem states that every solvable group occurs as a Galois group over ```Q```. Here, given a solvable group, via the index in the data base of small groups, and a bound on the discriminant, we find all normal field having this group as Galois group and of bounded discriminant.

The implementation is due to Carlo Sircana and Tommy Hofmann.
